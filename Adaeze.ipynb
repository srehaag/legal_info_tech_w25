```python
!pip install datasets -q

# Load data
from datasets import load_dataset
import pandas as pd
df = pd.DataFrame(load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir="RLLR"))

# Sort cases by most recent
df['document_date'] = pd.to_datetime(df['document_date'], format='mixed', errors='coerce')
df = df.sort_values(by='document_date', ascending=False)
df.head(5)
```

```python
from collections import Counter

def infer_pronouns(text):
    pronoun_categories = {
        'male': {'he', 'him', 'his'},
        'female': {'she', 'her', 'hers'},
        'other': {'they', 'their', 'theirs', 'them'}
    }

    words = text.lower().split()
    word_counts = Counter(words)

    category_counts = {category: sum(word_counts[word] for word in pronouns)
                       for category, pronouns in pronoun_categories.items()}

    return max(category_counts, key=category_counts.get, default='other')


df['gender'] = df['unofficial_text'].apply(infer_pronouns)
df.head(5)
```

```python
def find_outcome_in_text(text):
    # Sample keywords for outcomes, modify based on actual text patterns
    if 'granted' in text.lower():
        return 'Granted'
    elif 'denied' in text.lower():
        return 'Denied'
    if 'approved' in text.lower():
        return 'Approved'
    elif 'rejected' in text.lower():
        return 'Rejected'
    else:
        return 'Unknown'

df['outcome'] = df['unofficial_text'].apply(find_outcome_in_text)
df.head(5)
```

### Conclusions

(1) Gender cannot be reliably inferred from unofficial_text, because many of the cases are written in second perons (i.e. "You have claimed refugee protection because...). Instead, try generative AI or some other categorization system.

(2) It doesn't make sense to try to get to outcomes from words like granted, dismissed, because those words can appear for any number of reasons unconnected to the outcome. Instead, try generative AI or some other categorization system. In this dataset, each case is positive, so if your categorization was working it would say Granted for all of them.


```python
# Here is an example of how you could use generative AI

# set up openai

from openai import OpenAI

# get openai key from colab
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

# set up client
client = OpenAI()

# set up get_completion function
def get_completion(user_message,
        system_message="You are a helpful assistant to a Canadian law student",
        model = "gpt-4o",
        temperature = 0):

    completion = client.chat.completions.create(
        model=model,
        temperature=temperature,
        messages=[
            {"role": "system", "content": system_message},
            {"role": "user", "content": user_message}
        ]
    )
    return completion.choices[0].message.content

# test system
question = "Is Roncarelli still good law?"
print(get_completion(question))


```

```python
# setup process to create prompt

system_message = """You are an assistant to a refugee law lawyer.
You are given instructions and a text, and you extract the requested
information from the text. You only return the requested information,
without any explanation. If the answer is unclear, you rertun only
the text string: unclear"""

def generate_user_prompt(task, document):
  prefix = "INTRUCTIONS: " + task + "\n DOCUMENT: "
  suffix = "\n ANSWER: "
  return prefix + document + suffix

```

```python
# try to get gender

def get_gender(document):
  task = "Return the gender of the principal claimant. Return only male, female, or other."
  user_message = generate_user_prompt(task, document)
  return get_completion(user_message=user_message, system_message=system_message)

# filter df to only get 5 cases to reduce costs for testing
df = df.head(5)

# apply get_country to df and put in new col: GENAI_country
df["GENAI_gender"] = df.unofficial_text.apply(get_gender)
df

```

```python
# try to get outcome

def get_outcome(document):
  task = "Return outcome of the principal claim. Return only granted, denied, or other."
  user_message = generate_user_prompt(task, document)
  return get_completion(user_message=user_message, system_message=system_message)

# filter df to only get 5 cases to reduce costs for testing
df = df.head(5)

# apply get_outcome to df and put in new col: GENAI_outcome
df["GENAI_outcome"] = df.unofficial_text.apply(get_outcome)
df
```
